# Frequency Distribution

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import gutenberg
from nltk.probability import FreqDist

sample = gutenberg.raw('C:\\Users\VINAY KUMAR REDDY\\AppData\\Roaming\\nltk_data\\corpora\\gutenberg\\carroll-alice.txt')

word_tokens = word_tokenize(sample)
fdist = FreqDist(word_tokens)
print(fdist)
print(fdist.most_common(20)) # returns list of tuples (word,number of times it repeats in the given sample)
```

```python
fdist.plot(20)
```

# StopWords

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english')) #using set makes the result of words in ascending order present in English text file
print(stop_words) 
```

To Remove English Stopwords

```python
from nltk.tokenize import sent_tokenize,word_tokenize

filtered_words = []

sample = gutenberg.raw('C:\\Users\VINAY KUMAR REDDY\\AppData\\Roaming\\nltk_data\\corpora\\gutenberg\\carroll-alice.txt')
print(type(sample))

word_tokens = word_tokenize(sample)

for w in word_tokens:
    if w not in stop_words:
        if len(w)>3:
            filtered_words.append(w)

fdist = FreqDist(filtered_words)
print(fdist.most_common(20))
```

```python
fdist.plot(20)
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import genesis

sample = genesis.raw('C:\\Users\\VINAY KUMAR REDDY\\AppData\\Roaming\\nltk_data\\corpora\\genesis\\english-kjv.txt')
word_tokens = word_tokenize(str(sample)) # converting sample obj to String type because genesis.raw() throwing bytes class

filtered_words = [w for w in word_tokens if w not in stop_words and len(w)>3]
fdist = FreqDist(filtered_words)
print(fdist.most_common(20))
```

```python
fdist.plot(20)
```

# Bigrams - Two consecutive words that occur in a text

```python
from nltk.corpus import webtext, stopwords
from nltk import bigrams

text_words = [w.lower() for w in webtext.words('pirates.txt') if w not in stop_words and len(w)>3]
print(text_words)
```

```python
bigram_words = bigrams(text_words)

fdist = FreqDist(bigram_words)
fdist.most_common(20)
```

# Trigrams - Three consecutive words that occur in a text

```python
from nltk.corpus import webtext, stopwords
from nltk import trigrams

text_words = [w.lower() for w in webtext.words('grail.txt') if w not in stop_words and len(w)>3]
print(text_words)
```

```python
trigram_words = trigrams(text_words)

fdist = FreqDist(trigram_words)
fdist.most_common(20)
```

# ngrams - N continuous words in a text

```python
from nltk.corpus import gutenberg, stopwords
from nltk import ngrams

text_words = [w.lower() for w in gutenberg.words('shakespeare-hamlet.txt') if w not in stop_words and len(w)>3]
print(text_words)
```

```python
ngram_words = ngrams(text_words,5) # needs to mention number of grams

fdist = FreqDist(ngram_words)
fdist.most_common(20)
```

```python
fdist.plot(20)
```

# Stemming - Process for text normalization which involves reducing words to their root/base words from their derivational affixes or derived words

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import state_union

ps = PorterStemmer()

text_words = word_tokenize(state_union.raw('1961-Kennedy.txt'))

#example_text = """Driver a self driving carmight not be fun for a professional driver, but for a lazy driver, a self driving car saves a lot of driving"""
#text_words = word_tokenize(example_text)

# for "message", "President" stemmed words do not have any meaning so we go with lemmatization
for w in text_words:
    print(ps.stem(w))
```

# Lemmatization - which involves reducing words to their root/base words using vocabulary and morphological analysis. Lemmatization brings context to the words which is not done in stemming

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize('corpora')) # returns Corpus
print(lemmatizer.lemmatize('better', pos='a')) # 'a' for adjective
print(lemmatizer.lemmatize('best', pos='a'))
```

# POS Tagging - task to tag/label each word in a sentence with its grammatical group such as noun, pronoun, verb, adverb, adjective, tense and many more

CC - Coordinating Conjuction
JJ - Adjective
IN - Preposition / Subordinating Conjuction
JJR - Adjective, Comparative
JJS - Adjective, Superlative
NN - Nooun, Singular
NNP - Proper Noun, Singluar
PRP - Personal Pronoun

```python
from nltk import pos_tag

sample = """Google has actively been purging malware-laden apps from Play Store, but the company
has been having a hard time with it,
especially when some very popular apps are found to be complicit."""

tokenized_text = word_tokenize(sample)
print(tokenized_text)
pos_tag(tokenized_text)
```

# Named Entity Recognition with NLTK

To identify important named entities in test such as people, places, organizations, locations, dates, states, work of art etc.,

ORGANIZATION - FaceBook, Google 
PERSON - George W.Bush, Obama
LOCATION - 22 West Westin st, Mount Everest
DATE - July, 2019-05-2
GPE - India, Ukraine
TIME - three forty pm, 3:30am
MONEY - 100 million dollars, GBP 50.60
FACILITY - Washington Monument, Stonehenge

```python
from nltk import ne_chunk
from nltk import pos_tag
from nltk.tokenize import word_tokenize
from nltk.corpus import state_union

sample = state_union.raw('2005-GWBush.txt')
pos_words = pos_tag(word_tokenize(sample))

ner_words = ne_chunk(pos_words)
print(ner_words)
```

# Tokenization with Spacy

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp("""Google Maps' Street View team tracks locations across the world, uploading the images to its website.
        Some have been deemed the scariest, or creepiest, sites ever - incuding a sun-scroched field in North Carolina.""")
```

```python
for token in doc:
    print(token.text)
```

# Lemmatization with spacy

```python
for lemma in doc:
    print(lemma.text, lemma.lemma_)
```

# POS Tagging with spacy

```python
for token in doc:
    print(token.text,'----',token.pos_,'----',token.tag_)
```

To use visualizers for pos tags:

# Named Entity Recognition using spacy

```python
doc = nlp("""Tokyo is the largets city in the World, if the entire Tokyo metro area is included,
with a total of more than 38 million residents. Another Japanese city, Osaka, also has a very large population of almost 
20.5 million.""")

for ent in doc.ents:
    print(ent.text,'-----',ent.label_)
```
