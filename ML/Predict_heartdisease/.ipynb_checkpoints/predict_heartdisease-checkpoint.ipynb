# Data Processing

```python
# to get the columns having categorical values
for i in data:
    if len(data[i].value_counts()) <= 5 and i != 'target':
        print(i, end=",")
```

```python
dataset = pd.get_dummies(data, columns=['sex','cp','fbs','restecg','exang','slope','ca','thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardscaler = StandardScaler()
columns_to_scale = ['age','trestbps','chol','thalach','oldpeak']
dataset[columns_to_scale] =standardscaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.columns
```

```python
dataset.head() # 5 columns are scaled remaining columns are onehotencoded(get_dummies) {keeping untouched 'target' column}
```

```python
y = dataset['target']
X = dataset.drop(['target'],axis=1) 
```

```python
X.head()
```

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_class = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(knn_class, X, y, cv=10) # cv tells how many experiments
    knn_scores.append(score.mean())
```

knn_scores list tells us the accuracy will be from one of those elements.

```python
knn_scores # if we append only score, then elemets in each array appended to the score_list count depends on cv provided to Classifier
```

```python
plt.figure(figsize=(10,10))
plt.plot([k for k in range(1,21)], knn_scores, color='red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([k for k in range(1,21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier Scores for diff K values')
```

By looking into the knn_scores graph we are going to finalize the k=12 as our best neighbor count our accuracy will be around 85%

```python
knn_class = KNeighborsClassifier(n_neighbors=12) # so we are again hitting our classifier with k=12
score = cross_val_score(knn_class, X, y, cv=10)
```

```python
score.mean() # this is the max values from knn_scores list. we are finding for which k value we get this score using graph
```
